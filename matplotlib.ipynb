```python
#Customization of plots
years = [2014, 2015,2016, 2017,
        2018, 2019, 2020, 2021]
income = [55, 56, 62, 61, 
         72,72,73, 75]
income_ticks = list(range(50,81,2))

plt.plot(years, income)
plt.title("Income of John(in USD)", fontsize = 22, fontname ="Arial")
plt.xlabel("Year")
plt.ylabel("Yearly Income in USD")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])

plt.show()
```

```python
#Subplots
x = np.arange(100)
fig, axes = plt.subplots(2,2)
axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title("Sine Wave")

axes[0,1].plot(x, np.cos(x))
axes[0,1].set_title("Cosine Wave")

axes[1,0].plot(x, np.random.random(100))
axes[1,0].set_title("Random Fucntion")

axes[1,1].plot(x, np.log(x))
axes[1,1].set_title("Log Wave")
axes[1,1].set_xlabel("X")

fig.suptitle("Four Plots")
plt.show()




```

```python
#3D Plotting
ax = plt.axes(projection = "3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x,y,z)
ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```

```python
#3D Plotting
ax = plt.axes(projection = "3d")

x = np.arange(0,50,0.1)
y = np.arange(0,50,0.1)
z = np.cos(x+y)

ax.plot(x,y,z)
ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```

```python
ax = plt.axes(projection = "3d")

x = np.arange(-5,5, 0.1)
y = np.arange(-5,5,0.1)
X, Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap = "Spectral")
ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```

```python
ax = plt.axes(projection = "3d")

x = np.arange(-5,5, 0.1)
y = np.arange(-5,5,0.1)
X, Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap = "Spectral")
ax.set_title("3D Plot")
ax.set_xlabel("test")
ax.view_init(azim = 0, elev = 90)
plt.show()
```
